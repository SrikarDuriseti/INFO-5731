```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('punkt')

sample='This is a great little speaker! You get hours of play time and the sound is even better than I expected. The size is great, you can literally throw it in your pocket or clip it to a bag. It’s very easy to set up and use. I got it for when I go paddle boarding, but it works so well I use it all the time now.'
```

```python
# tokenization
tokenized_text= nltk.word_tokenize(sample)
print(tokenized_text)

#bag of words

vector=CountVectorizer()
vectorized_sample=vector.fit_transform([sample]) #fitting the data to the object
vect_array=vectorized_sample.toarray()
feature_names=vector.get_feature_names_out()  #features
print('\nBAG OF WORDS\n')
for i, count in zip(feature_names, vect_array[0]):
    print(i,"-",count)

#Tf-Idf
tf_idf=TfidfVectorizer()
tf_idf_vectors=tf_idf.fit_transform([sample]) #fit and transform the review
feature_names=tf_idf.get_feature_names_out() #extracting the features out

print('\nTFIDF Vectorization\n')
tf_idf_score={}

for i in tf_idf_vectors.nonzero()[1]:
    tf_idf_score[feature_names[i]] = tf_idf_vectors[0,i]

for word, score in sorted(tf_idf_score.items(), key=lambda x: x[1], reverse=True):   #generate=ing the score for each word
    print(f"{word}: {score:.3f}")
```

Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

```python
# breaking down the sample data into list in four sentences so that comparison and mapping with the words can be done

sample='This is a great little speaker! You get hours of play time and the sound is even better than I expected. The size is great, you can literally throw it in your pocket or clip it to a bag. It’s very easy to set up and use. I got it for when I go paddle boarding, but it works so well I use it all the time now.'

sample_list=['This is a great little speaker! You get hours of play time and the sound is even better than I expected',
             ' The size is great, you can literally throw it in your pocket or clip it to a bag.',
             ' It’s very easy to set up and use.',
             ' I got it for when I go paddle boarding, but it works so well I use it all the time now.']
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

labels = ['expected','throw','easy','boarding'] #four

label_dict = {word: index for index,word in enumerate(labels)}
label_num = [label_dict[i] for i in labels]

#building tf-idf vectorizer
tf=TfidfVectorizer()
new_tf=tf.fit_transform(sample_list)
```

```python
# Random Forest classifier
rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(new_tf, label_num)

#dictionary created out of features and its importance
impt_features = rf.feature_importances_
#dataframe with the features extracted and its importance
df = pd.DataFrame({'Features': tf.get_feature_names_out(),'Importance': impt_features})
df
```

```python
final_df=df.sort_values(by='Importance',ascending=False)
final_df
```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
modelname = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(modelname)
model = AutoModel.from_pretrained(modelname)

query = "Go through the manual"

sample_text= [" Little did she know, it was not just tricks but real magic at play",
             "As the show reached its climax, the magician whispered a secret spell",
             "suddenly, the girl found herself floating above the stage, surrounded by shimmering lights",
             " Gasps filled the air as the audience witnessed the impossible",
             " It was a moment of pure enchantment, where reality blurred with fantasy. "]

#Using the BERT transformer embedding the text
def get_bert_embeddings(text):
    inp = tokenizer(text, return_tensors="pt",truncation=True,padding=True)
    with torch.no_grad():
        output = model(**inp) #output with the input that used a tokenizer
    embeddings = output.last_hidden_state.mean(dim=1).squeeze().numpy()
    return embeddings

def rank_text_similarity(query_embeddings, sample_embeddings, sample_text):
    similarities = cosine_similarity([query_embeddings], sample_embeddings)[0]
    ranks = sorted(enumerate(similarities), key=lambda x: x[1], reverse=True)
    result = [(sample_text[index], similarity) for index, similarity in ranks] # similarity and its rank is the output
    return result

# BERT embeddings into query and sample text data
query_embeddings = get_bert_embeddings(query)
sample_embeddings = [get_bert_embeddings(i) for i in sample_text]

# Rannks based on similarity
ranked_similarity_result = rank_text_similarity(query_embeddings, sample_embeddings, sample_text)

print("Ranked based on Similarity to the query ")
for rank, (text, similarity) in enumerate(ranked_similarity_result):
    print(f"Rank {rank + 1}: Similarity = {similarity:.4f}")
    print(f"Text: {text}\n")
```
